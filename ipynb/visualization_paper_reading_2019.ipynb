```python
lines = text.split('\n')

titles = []
for line in lines:
    line = line.replace('- [', '')
    line = line.split('](')[0]
    line = line.replace(':', ' ')
    line = line.replace('(', ' ')
    line = line.replace(')', ' ')
    if line.strip() and not line.startswith('#'):
#         print(line)
        titles.append(line)
```

```python
venues = []
for line in lines:
#     print(line)
    if 'kbd' not in line:
        continue
    line = line.split('<kbd>')[1]
    line = line.split('</kbd>')[0]
    line = line.replace('(', '')
    line = line.replace(')', '')
    line = line.replace('IEEE', ' ')
    line = line.replace('oral', ' ')
    line = line.replace('Talk', ' ')
    line = line.replace('Spotlight', ' ')
    line = line.replace('2014 IV', 'IV 2014')
    line = line.strip()
    if len(line.split()) == 1:
        continue
    if line and not line.startswith('#'):
#         print(line)
        venues.append(line)
```
